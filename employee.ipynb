```python
import sqlite3
import pandas as pd 
```

```python
def create_dbase(dbase_name,file_name,table_name):
    con=sqlite3.connect(dbase_name)
    cursor=con.cursor()
    employee_data=pd.read_csv(file_name)
    employee_data.to_sql(table_name,con,if_exists='replace',index=False)

    results=cursor.execute(f"select *from {table_name}").fetchall()
    con.commit()
    con.close()

    return results
```

```python
employee_data=create_dbase("Employee.db","Employee.csv","Employee")
```
